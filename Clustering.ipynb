```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pickle

# Load your dataset
data = pd.read_csv("the_og_processed_data.csv")

# Selecting only numerical columns for clustering
num_cols = data.select_dtypes(include=[np.number]).columns
X = data[num_cols]

# Normalizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Finding the optimal number of clusters using the Elbow Method
inertia = []
silhouette_scores = []
K_range = range(2, 5)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

optimal_k = K_range[np.argmax(silhouette_scores)]
print(f"Optimal K: {optimal_k}")

# Train the final model
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans.fit(X_scaled)

# Save the model
with open("the_og_clustering_model.pkl", "wb") as f:
    pickle.dump((kmeans, scaler), f)
print("done")
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap="viridis", alpha=0.5)
plt.colorbar()
plt.title("Clustering Results with K-Means")
plt.show()
```

```python
import numpy as np

print("Unique clusters found:", np.unique(kmeans.labels_))
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Get cluster labels
labels = kmeans.labels_

# Plot each cluster separately with different colors
plt.figure(figsize=(8, 6))
for cluster in np.unique(labels):
    plt.scatter(
        X_pca[labels == cluster, 0], 
        X_pca[labels == cluster, 1], 
        label=f"Cluster {cluster}", 
        alpha=0.5
    )

# Plot cluster centers
centers_pca = pca.transform(kmeans.cluster_centers_)  # Transform centroids to 2D
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', marker='X', s=200, label='Centroids')

plt.legend()
plt.title("Improved Clustering Visualization")
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
for cluster in np.unique(labels):
    plt.scatter(
        X_tsne[labels == cluster, 0], 
        X_tsne[labels == cluster, 1], 
        label=f"Cluster {cluster}", 
        alpha=0.5
    )

plt.legend()
plt.title("Clustering Visualization using t-SNE")
plt.show()
```
